# Limpieza básica de encoding (ftfy)

Este notebook:
- Carga un archivo Excel
- Corrige texto con problemas de encoding usando `ftfy`
- Exporta un CSV limpio compatible con Excel

```python
import pandas as pd
import ftfy
```

```python
df = pd.read_excel("../data/raw/data_train.xlsx")
df.head()
```

```python
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].apply(
        lambda x: ftfy.fix_text(x) if isinstance(x, str) else x
    )
```

```python
df.to_csv("../data/processed/data_fixed_v2.csv", index=False, encoding="utf-8-sig")

print("Archivo guardado en ../data/processed/data_fixed_v2.csv")
```
