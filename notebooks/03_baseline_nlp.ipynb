# Baseline NLP — TF-IDF + LinearSVC



```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.svm import LinearSVC

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    f1_score,
)
```

```python
df = pd.read_csv("../data/processed/data_cleaned_v2md.csv")

df.head()
```

```python
print(df.columns.tolist())
```

```python
X = df["tweet_text_clean"]

Y = df["class"]
```

```python
print(df["class"].value_counts())

print("\nDistribución porcentual:\n")

print(df["class"].value_counts(normalize=True))
```

```python
df[["tweet_text_clean", "class"]].sample(20, random_state=42)
```

```python
X_train, X_test, y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42,
    stratify=Y
)
```

```python
vectorizer = TfidfVectorizer(
    max_features=10000,
    ngram_range=(1,2),
    min_df=2
)

X_train_tfidf = vectorizer.fit_transform(X_train)

X_test_tfidf = vectorizer.transform(X_test)
```

```python
print("Train shape:", X_train_tfidf.shape)

print("Test shape:", X_test_tfidf.shape)
```

```python
model = LinearSVC()

model.fit(X_train_tfidf, y_train)
```

```python
y_pred = model.predict(X_test_tfidf)
```

```python
accuracy = accuracy_score(Y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
f1 = f1_score(Y_test, y_pred, average="weighted")

print("F1-score:", f1)
```

```python
print(classification_report(Y_test, y_pred))
```

```python
cm = confusion_matrix(Y_test, y_pred)

cm
```

```python
results = pd.DataFrame({
    "text": X_test,
    "real": Y_test,
    "predicted": y_pred
})

results.sample(20, random_state=42)
```

```python
errors = results[results["real"] != results["predicted"]]

errors.sample(min(20, len(errors)), random_state=42)
```
